Automat vypouští 80ml do lahví. Zaměstnanec tomu nevěří a tak vezme 40 vzorku a naměří 78ml +- 2.5. \
1) nulová a alternativní hypotéza?
2) s 95% spolehlivostí, je dost důkazů na potvrzení zaměstnancova tvrzení?

(1)
\begin{align*}
H_0: \mu = 80 ml \newline
H_A: \mu \ne 80 ml
\end{align*}

(2)
- 80 ml != --> two tailed test

```python
nullHypothesisValue: int = 80
sampleSize: int = 40
sampleMean: int = 78
sampleStdev: float = 2.5
confidenceLevel: float = 0.95

if sampleSize > 30:
    print("Sample size > 30 and stdev is known, using normal distribution")
    from scipy.stats import norm
    alphaValue_twosided: float = (1+confidenceLevel)/2
    
    zValue: float = (sampleMean-nullHypothesisValue)/(sampleStdev/(sampleSize**0.5))
    print(zValue)
    zkrit: float = norm.ppf(alphaValue_twosided)
    
    
    isAccepted: bool = abs(zValue) < (abs(zkrit))
    if isAccepted:
        print(f"Z-value {zValue} of {nullHypothesisValue} ml is accepted, because it is lower than right z-crit {zkrit}")
        print("According to this statement, employee is false")
    else:
        print(f"Z-value {zValue}  of {nullHypothesisValue} ml is rejected, because it is higher than right z-crit  {zkrit}")
        print("According to this statement, employee is right")
    
    
    ErrorBoundMean: float = zkrit * (sampleStdev/(sampleSize)**0.5)

    print(f"z = {zValue}")
    print(f"zkrit = {zkrit}")
    print(f"EBM = {ErrorBoundMean}")
    print()

    rejectionLimit: float = sampleMean + ErrorBoundMean
    isAccepted: bool = nullHypothesisValue <= (rejectionLimit)
    if isAccepted:
        print(f"Null hypothesis value of {nullHypothesisValue} ml is accepted, because it is lower than right z-crit with absolute value {rejectionLimit}")
        print("According to this statement, employee is false")
        
    else:
        print(f"Null hypothesis value of {nullHypothesisValue} ml is rejected, because it is higher than right z-crit with absolute value {rejectionLimit}")
        print("According to this statement, employee is right")




else:
    print("sampleSize < 30, using student distribution")
    from scipy.stats import t
    degreesOfFreedom: int = sampleSize-1
    significance: float = 1 - confidenceLevel
    alphaValue_twosided: float = (1+confidenceLevel)/2

    tvalue: float = t.ppf(alphaValue_twosided, df = degreesOfFreedom)
    ErrorBoundMean: float = t * (sampleStdev/(sampleSize)**0.5)

    print(f"t = {t}")
    print(f"EBM = {ErrorBoundMean}")
    print()

    rejectionLimit: float = sampleMean + ErrorBoundMean
    isAccepted: bool = nullHypothesisValue <= (rejectionLimit)
    if isAccepted:
        print(f"Null hypothesis value of {nullHypothesisValue} ml is accepted, because it is lower than right z-crit with absolute value {rejectionLimit}")
        print("According to this statement, employee is false")
    else:
        print(f"Null hypothesis value of {nullHypothesisValue} ml is rejected, because it is higher than right z-crit with absolute value {rejectionLimit}")
        print("According to this statement, employee is right")
```
