Když mám několik vzorků a chci zjistit, jestli jsou jejich průměry stejné, používá se tzv f-statistika.\
- z-test: jeden člen z populace
- z-statistika: průměr vzorku z populace, a mám smodch populace
- t-statistika: průměr vzorku z populace, nemám smodch populace (málo vzorků)
- f-statistika: ...

SST: sum of squares total, vezmu součet rozdílů VŠECH hodnot VŠECH VZORKŮ vůči GLOBÁLNÍMU PRŮMĚRU\
Má m·n - 1 DOF
- m: počet vzorkových "populací"
- n: počet vzorků v populacích, měl by být stejný


\
\
\
\
\

Mám  3 vzorky z populace:

- 3, 2, 1
- 5, 3, 4
- 5, 6, 7
\
\
\
\
--> nTot = 9\
--> n = 3\
--> průměr 1 (x1_) = (3+2+1)/3 = 2\
--> průměr 2 (x2_) = (5+3+4)/3 = 4\
--> průměr 3 (x3_) = (5+6+7)/3 = 6\
--> průměr celkový (xt_) = ((3+2+1) + (5+3+4) + (5+6+7))/9 = 4

\
\
\
\

SST: Sum of squares total, celkový rozptyl hodnot

\begin{align*}
SST = SST + SSW\newline
SST = \sum{\sum{(x_i - \bar{x_j})^2}} \newline
DOF_{SST} = m·n-1 \newline
\end{align*}
i ... index v rámci vzorku j\
j ... vzorková populace\
_x.j ... průměr populace j\

SST = [(3-4)^2 + (2-4)^2 + (1-4)^2] + [(5-4)^2 + (3-4)^2 + (4-4)^2] + [(5-4)^2 + (6-4)^2 + (7-4)^2] = 30\
SST_DOF = m·n-1 = 3·3 - 1 = 8
\
\
\
\
\
\
\
\
\

\
\
SSW: sum of squares within samples, kolik z celkové variance dělají variance vzorků vůči svým korespondujícím průměrům

\begin{align*}
SSW = \sum{(x_i - \bar{x})^2} \newline
DOF_{SSW} = m·(n-1) \newline
\end{align*}

i ... index člena všech vzorků v jednom arrayi\
_x ... celkový průměr\

SSW = [(3-2)^2 + (2-2)^2 + (1-2)^2] + [(5-4)^2 + (3-4)^2 + (4-4)^2] + [(5-6)^2 + (6-6)^2 + (7-6)^2] = 6\
SSW_DOF = 3·(3-1) = 6
\
\
\
\
\

\
\
\
SSB: sum of squares between samples, kolik z celkové variance dělají variance vzorků vůči sobě navzájem\
Jednodušeji: místo každé hodnoty každého vzorku dávám průměr daného odečtený od celkového průměru

\begin{align*}
SSB = \sum{\sum{(\bar{x_j} - \bar{x})^2}} \newline
DOF_{SSB} = m - 1 \newline
\end{align*}

SSW = [(2-4)^2 + (2-4)^2 + (2-4)^2] + [(4-4)^2 + (4-4)^2 + (4-4)^2] + [(6-4)^2 + (6-4)^2 + 6-4)^2] = 24\
SSB_DOF = 3-1 = 2
\
\
\


\begin{align*}
SST = SSW + SSB \newline
DOF_{SST} = DOF_{SSW} + DOF_{SSB} 
\end{align*}

SST = 30, SSW = 6, SSB = 24, 30 = 6 + 24 \
DOF_SST = 8, DOF_SSW = 6, DOF_SSB = 2, 8 ==  6 + 2\
\
\
\
\

f-statistika: dle biologa Fischera\
opět má nějakou distribuci pro scipyho\

\begin{align*}
f = \frac{\frac{SSB}{DOF_{SSB}}}{\frac{SSW}{DOF_{SSW}}} = \frac{\frac{SSB}{m-1}}{\frac{SSW}{m·(n-1)}}
\end{align*}

Co to říká?\
Když je čitatel mnohem vyšší než jmenovatel, tak je celková variance dat daleko více ovlivněná variancí mezi průměry, ne variancí dat mezi vzorky\
\to by naznačovalo, že naše průměry pravděpodobně nereprezentují reálný průměr populace
\
\
Když je jmenovatel mnohem vyšší než čitatel, tak je to naopak a znamená to, že variace průměrů mezi vzorky je asi spíše náhodná.
\
\
Jinak se na f-distribuci mohu dívat jako na poměr dvou chí-kvadrát distribucí, což vidím i tady na f-statistice.
